```python
def check(actual, expected):
  if expected != actual:
    print(f"Function should return the value {expected}, it is returning the value {actual}")
  else:
    print(f"Congratulations, the test case passed!")


def correct_neighbours(G, x, y):
  return y in G[x]


def correct_almost_neighbours(G, x, y):
  if x == y or correct_neighbours(G, x, y):
    return False
  for i in G[x]:
    if y in G[i]:
      return True
  return False


def correct_solve(n):
  G = [[] for i in range(n + 1)]
  if n % 2 == 0:
    for i in range(1, n + 1):
      for j in range(1, n + 1):
        if i != j and i + j != n + 1:
          G[i].append(j)
  else:
    for i in range(1, n):
      for j in range(1, n):
        if i != j and i + j != n:
          G[i].append(j)
    for i in range(1, n):
      G[i].append(n)
      G[n].append(i)
  return G


def correct_dfs(v, s, G, u):
  u[v] = True
  for i in G[v]:
    if i != s and not u[i]:
      correct_dfs(i, s, G, u)


def correct_count_important(G):
  n = len(G)
  answer = 0
  for s in range(n):
    u = [False] * n
    if s == 0:
      correct_dfs(1, s, G, u)
    else:
      correct_dfs(0, s, G, u)
    cnt = 0
    for i in range(n):
      cnt += u[i]
    answer += cnt != n - 1
  return answer


def correct_shortest_distance(G, source):
  n = len(G)
  distance = [n] * n
  distance[source] = 0
  q = [source]
  l = 0
  while l != len(q):
    v = q[l]
    l += 1
    for to in G[v]:
      if distance[to] > distance[v] + 1:
        distance[to] = distance[v] + 1
        q.append(to)
  return distance


def correct_dfsPlus(v, p, d, G, used, up, depth):
  used[v] = True
  up[v] = d
  depth[v] = d
  vis = 0
  ans = 0
  for to in G[v]:
    if not used[to]:
      ans += correct_dfsPlus(to, v, d + 1, G, used, up, depth)
      vis += 1
      up[v] = min(up[v], up[to])
    elif to != p:
      up[v] = min(up[v], depth[to])
  if d == 0:
    if vis > 1:
      ans += 1
    return ans
  if len(G[v]) > 1 and up[v] == depth[v]:
    ans += 1
  return ans


def correct_count_important_fast(G):
  n = len(G)
  used = [False] * n
  depth = [0] * n
  up = [0] * n
  return correct_dfsPlus(0, -1, 0, G, used, up, depth)
```
